```python
songs = songs[['id','name','valence','acousticness','danceability','energy','duration_ms','explicit','instrumentalness','key','popularity','release_date','speechiness','tempo','artists','year']]
```

```python
songs = songs.drop_duplicates(subset=['name','year'])
```

```python
def scalc(val1,val2):
    return min(val1,val2)/max(val1,val2)

def similarity(songid1,songid2):
    a = songs.iloc[songid1]
    b = songs.iloc[songid2]
    
    valenceA = a['valence']
    valenceB = b['valence']
    valencesim = scalc(valenceA,valenceB)
    
    acousticnessA = a['acousticness']
    acousticnessB = b['acousticness']
    acousticnesssim = scalc(acousticnessA,acousticnessB)
    
    danceabilityA = a['danceability']
    danceabilityB = b['danceability']
    danceabilitysim = scalc(danceabilityA,danceabilityB)
    
    energyA = a['energy']
    energyB = b['energy']
    energysim = scalc(energyA,energyB)
    
    instrumentalnessA = a['instrumentalness']
    instrumentalnessB = b['instrumentalness']
    instrumentalnesssim = scalc(instrumentalnessA,instrumentalnessB)
    
    speechinessA = a['speechiness']
    speechinessB = b['speechiness']
    speechinesssim = scalc(speechinessA,speechinessB)
    
    tempoA = a['tempo']
    tempoB = b['tempo']
    temposim = scalc(tempoA,tempoB)
    
    value = (valencesim+acousticnesssim+danceabilitysim+energysim+instrumentalnesssim+speechinesssim+temposim)/7    
    return value
```

```python
def most_similar_finder(songid):
    max_similarity = -1
    most_similar_songid = -1
    for i in range(len(songs)):
        if(i == songid): 
            pass
        else:
            s = similarity(songid,i)
            max_similarity = max(max_similarity,s)
            if(max_similarity == s): most_similar_songid = i
    return most_similar_songid        
```

```python
song_name = input()
song_id = -1
for i in range(len(songs)):
    if(songs.iloc[i]['name'] == song_name): song_id = i; break  
        
if(song_id == -1): print("No similar song found in the dataset")
else:
    val = most_similar_finder(song_id)
    similar_song = songs.iloc[val]
    Name = similar_song['name']
    print(f'Most Similar Song Found in the DataSet is: {Name}')
    print(f'Similarity %: {similarity(val,song_id)*100}')
    artist = similar_song['artists']
    print(f'Artists: {artist}')
```
